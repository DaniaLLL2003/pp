### Задание 2

```python
t = np.linspace(0, 10, 51)
t
```

```python
f = np.cos(t)
f
```

```python
diagram = pd.Series(f, t)
diagram
```

```python
plt.axis([0.5, 9.5, -2.5, 2.5])

label_dict = {'fontsize':'14', 'color':'green', 'family':'Calibri'}

diagram.plot(color='green')

plt.title('График f(t)', fontdict=label_dict)

plt.xlabel('Значения t')
plt.ylabel('Значения f')
```

### Задание 3

```python
x = np.linspace(-3, 3, 51)
x
```

```python
x1 = x**2
x2 = 2 * x + 0.5
x3 = -3 * x - 1.5
x4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
y1, y2, y3, y4 = ax.flatten()

y1.plot(x, x1)
y1.set_title('График y1')
y1.axis([-5, 5, -1, 9])

y2.plot(x, x2)
y2.set_title('График y2')

y3.plot(x, x3)
y3.set_title('График y3')

y4.plot(x, x4)
y4.set_title('График y4')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
t = df['Class'].value_counts()
t
```

```python
df_class_info = pd.Series(t)
df_class_info.plot(kind='bar', logy=True)
plt.show()
```

```python
v1_class1=df.set_index('Class')['V1'].filter(like='1', axis=0)
v1_class1=v1_class1.reset_index()
v1_class1=v1_class1.drop('Class', axis=1)
v1_class1.head()
```

```python
v1_class0=df.set_index('Class')['V1'].filter(like='0', axis=0)
v1_class0=v1_class0.reset_index()
v1_class0=v1_class0.drop('Class', axis=1)
v1_class0.head()
```

```python
plt.hist(v1_class0['V1'], bins=20, color='grey', density = True, ec='black', orientation='horizontal', alpha=0.5)
plt.hist(v1_class1['V1'], bins=20, color='red', density = True, ec='black', orientation='horizontal', alpha=0.5)
plt.plot()
plt.xlabel('Class')
plt.legend(labels=['Class 0', 'Class 1'])
```
